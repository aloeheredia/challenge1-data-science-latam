### Importación de datos



```python
import pandas as pd
from matplotlib import pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 6. Visualización

6.1. Ingresos por tienda

```python
# Datos
tienda = ["Tienda 4","Tienda 3", "Tienda 2", "Tienda 1"]
ingresos = [suma4,suma3,suma2,suma1]

# Crear el gráfico de barras horizontales
bars = plt.barh(tienda, ingresos, color='skyblue')

# Formatear el eje x
plt.ticklabel_format(style='plain', axis='x')

# Rotar las etiquetas del eje x y ajustar alineación
plt.xticks(rotation=45, ha='right')

# Agregar etiquetas a cada barra
plt.bar_label(bars, labels=ingresos, label_type='edge')

# Ajustar márgenes para que quepan las etiquetas
plt.subplots_adjust(bottom=0.3, left=0.2)
# Personalizar
plt.xlabel("Ingresos")
plt.ylabel("Tienda")
plt.title("Ingresos por Tienda")
plt.grid(axis='x', linestyle='--', alpha=0.6)  # Líneas de referencia verticales

# Mostrar
plt.show()
```

6.2. Ventas por Categorías

```python
#TIENDA 1
lista_categoria1 = []
lista_venta_categoria1 = []
for categoria in tienda['Categoría del Producto'].unique():
  ventas_categoria1= tienda[tienda['Categoría del Producto'] == categoria]['Precio'].sum()
  lista_categoria1.append(categoria)
  lista_venta_categoria1.append(float(ventas_categoria1))
# Crear una figura más grande
plt.figure(figsize=(12, 6))

# Crear el gráfico de barras
bars = plt.bar(lista_categoria1, lista_venta_categoria1)

# Formatear el eje y
plt.ticklabel_format(style='plain', axis='y')

# Rotar las etiquetas del eje x y ajustar alineación
plt.xticks(rotation=45, ha='right')

# Agregar etiquetas a cada barra
plt.bar_label(bars, labels=lista_venta_categoria1, label_type='edge')

# Ajustar márgenes para que quepan las etiquetas
plt.subplots_adjust(bottom=0.3)

# Títulos y etiquetas
plt.title("Ventas por Categoría")
plt.xlabel("Categorías")
plt.ylabel("Ventas (S/)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.ylim(0, max(lista_venta_categoria1) * 1.1)

# Mostrar el gráfico
plt.show()
```
